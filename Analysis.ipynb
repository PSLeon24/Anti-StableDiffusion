```python
df = pd.read_csv('metrics.csv')

metrics_cols = ['Original_LPIPS', 'Adversarial_LPIPS', 'Original_SSIM', 'Adversarial_SSIM', 'Original_PSNR', 'Adversarial_PSNR']

# 1. Statistical Summary
print("Statistical Summary:")
print(df[metrics_cols].describe())

# 2. LPIPS (Original vs Adversarial) Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Original_LPIPS', 'Adversarial_LPIPS']])
plt.title("LPIPS (Original vs Adversarial) Comparison")
plt.ylabel("LPIPS Score")
plt.show()

# 3. SSIM (Original vs Adversarial) Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Original_SSIM', 'Adversarial_SSIM']])
plt.title("SSIM (Original vs Adversarial) Comparison")
plt.ylabel("SSIM Score")
plt.show()

# 4. PSNR (Original vs Adversarial) Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Original_PSNR', 'Adversarial_PSNR']])
plt.title("PSNR (Original vs Adversarial) Comparison")
plt.ylabel("PSNR Score")
plt.show()

# 5. LPIPS (Original vs Adversarial) Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Original_LPIPS', y='Adversarial_LPIPS', data=df, alpha=0.7)
plt.title("LPIPS (Original vs Adversarial) Correlation")
plt.xlabel("Original LPIPS")
plt.ylabel("Adversarial LPIPS")
plt.show()

# 6. SSIM (Original vs Adversarial) Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Original_SSIM', y='Adversarial_SSIM', data=df, alpha=0.7)
plt.title("SSIM (Original vs Adversarial) Correlation")
plt.xlabel("Original SSIM")
plt.ylabel("Adversarial SSIM")
plt.show()

# 7. PSNR (Original vs Adversarial) Scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Original_PSNR', y='Adversarial_PSNR', data=df, alpha=0.7)
plt.title("PSNR (Original vs Adversarial) Correlation")
plt.xlabel("Original PSNR")
plt.ylabel("Adversarial PSNR")
plt.show()

# 8. Pairplot for Metrics
sns.pairplot(df[metrics_cols], diag_kind='kde', plot_kws={'alpha': 0.5})
plt.suptitle("Relationship Between Metrics (Pairplot)", y=1.02)
plt.show()

# 9. Bar Plot for Mean Values of Metrics
#  Calculate mean values for each metric (Original vs Adversarial)
lpips_means = df[['Original_LPIPS', 'Adversarial_LPIPS']].mean()
ssim_means = df[['Original_SSIM', 'Adversarial_SSIM']].mean()
psnr_means = df[['Original_PSNR', 'Adversarial_PSNR']].mean()

# Create subplots for LPIPS, SSIM, and PSNR
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

# Bar plot for LPIPS
lpips_means.plot(kind='bar', ax=axes[0], color=['blue', 'orange'])
axes[0].set_title("Mean LPIPS (Original vs Adversarial)")
axes[0].set_ylabel("Mean Score")
axes[0].set_xticklabels(['Original_LPIPS', 'Adversarial_LPIPS'], rotation=45)
axes[0].set_ylim([0, max(lpips_means) * 1.1])  # y  

# Bar plot for SSIM
ssim_means.plot(kind='bar', ax=axes[1], color=['blue', 'orange'])
axes[1].set_title("Mean SSIM (Original vs Adversarial)")
axes[1].set_xticklabels(['Original_SSIM', 'Adversarial_SSIM'], rotation=45)
axes[1].set_ylim([0, max(ssim_means) * 1.1]) 

# Bar plot for PSNR
psnr_means.plot(kind='bar', ax=axes[2], color=['blue', 'orange'])
axes[2].set_title("Mean PSNR (Original vs Adversarial)")
axes[2].set_xticklabels(['Original_PSNR', 'Adversarial_PSNR'], rotation=45)
axes[2].set_ylim([0, max(psnr_means) * 1.1])  # y  

plt.tight_layout()
plt.show()


# 10. SSIM (Original vs Adversarial) Distribution - KDE Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Original_SSIM'], label='Original SSIM', shade=True)
sns.kdeplot(df['Adversarial_SSIM'], label='Adversarial SSIM', shade=True)
plt.title("SSIM (Original vs Adversarial) Distribution")
plt.xlabel("SSIM Score")
plt.legend()
plt.show()

# 11. LPIPS (Original vs Adversarial) Distribution - KDE Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Original_LPIPS'], label='Original LPIPS', shade=True)
sns.kdeplot(df['Adversarial_LPIPS'], label='Adversarial LPIPS', shade=True)
plt.title("LPIPS (Original vs Adversarial) Distribution")
plt.xlabel("LPIPS Score")
plt.legend()
plt.show()

# 12. PSNR (Original vs Adversarial) Distribution - KDE Plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Original_PSNR'], label='Original PSNR', shade=True)
sns.kdeplot(df['Adversarial_PSNR'], label='Adversarial PSNR', shade=True)
plt.title("PSNR (Original vs Adversarial) Distribution")
plt.xlabel("PSNR Score")
plt.legend()
plt.show()

# 13. Correlation Heatmap
plt.figure(figsize=(10, 8))
correlation = df[metrics_cols].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Between Metrics")
plt.show()
```
